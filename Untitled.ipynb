Artists with most songs

```python
df.artists.nunique() #There are 34088 artists in the entire dataset.
```

```python
df.artists.value_counts()[:7]
```

we first create a list using the index returned by the value_counts

```python
artist_list=df.artists.value_counts().index[:7]
```

The we will filter the dataframe using this top 7 artists and group by year

```python
df_artists=df[df.artists.isin(artist_list)][['energy','year','artists']].groupby(['artists','year']).count().reset_index()
df_artists.rename(columns={'energy':'song_count'},inplace=True)
```

```python
df_artists.head()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(x='year',y='song_count',data=df_artists,hue='artists')
```

I will now try a different way to see which artists are dominating which era. First, I will create an empty dataframe that contains the entire timeline (19212020) and the names of top 7 artists.

```python
df1=pd.DataFrame(np.zeros((100,7)),columns=artist_list)
df1['year']=np.arange(1921,2021)
df1.head()
```

```python
df1=df1.melt(id_vars='year',var_name='artists',value_name='song_count')
df1.head()
```

```python
df_merge=pd.merge(df1,df_artists,on=['year','artists'],how='outer').sort_values(by='year').reset_index(drop=True)
df_merge.head()
```

```python
df_merge.fillna(0,inplace=True)
df_merge.drop('song_count_x',axis=1,inplace=True)
df_merge.rename(columns={'song_count_y':'song_count'},inplace=True)
df_merge.head()
```

```python
df_merge['cumsum']=df_merge[['artists','song_count']].groupby('artists').cumsum()
df_merge.head(15)
```

```python
import plotly.express as px

fig=px.bar(data_frame=df_merge,x='artists',y='cumsum',color='artists',animation_frame='year',animation_group='year',
          range_y=[0,2000],title='Artists with Most Number of Songs')
fig.show()
```
